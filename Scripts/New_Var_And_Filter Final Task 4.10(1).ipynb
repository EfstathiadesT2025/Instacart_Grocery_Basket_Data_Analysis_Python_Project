```python
# 2) Addressing any PII data in the dataframe - omitting the 'Surname' column using vars_list
vars_list = ['order_id', 'user_id', 'order_number', 'order_dow', 'order_hour_of_day', 'days_since_prior_order', 'product_id', 'add_to_cart_order', 'reordered', '_merge1', 'product_name', 'aisle_id', 'department_id',
       'prices', '_merge2', 'Busiest days', 'busiest_period_of_day',
       'max_order', 'loyalty_flag', 'average_price', 'order_frequency',
       'First Name', 'Gender', 'STATE', 'Age', 'date_joined',
       'n_dependants', 'fam_status', 'income', '_merge']
```

```python
# 3) Customer behavior in different geographic regions
# 3a) New column 'Region' - assigning the string values
result = []

for value in df_merged_final2["STATE"]:
  if value in ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'Pennsylvania', 'New Jersey']:
    result.append("Northeast")
  elif value in ['Wisconsin', 'Michigan', 'Illinois', 'Indiana', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri']:
    result.append("Midwest")
  elif value in ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana']:
    result.append("South")
  elif value in ['Idaho', 'Montana', 'Wyoming', 'Nevada', 'Utah', 'Colorado', 'Arizona', 'New Mexico', 'Alaska', 'Washington', 'Oregon', 'California', 'Hawaii']:
    result.append("West")
  else:
    result.append("No region")
```

```python
result
```

```python
# Creating the new column within the dataframe and merging it with the result values
df_merged_final2['Region'] = result
```

```python
# Frequency check
df_merged_final2['Region'].value_counts(dropna = False)
```

```python
# 4) Creating an exclusion flag for low-activity customers (customers with less than 5 orders)
result2 = []

for value in df_merged_final2['order_number']:
    if value < 5:
        result2.append('Low-activity customer')
    elif 10 < value >= 5:
        result2.append('Regular customer')
    else:
        result2.append('No orders')
```

```python
result2
```

```python
df_merged_final2['exclusion_flag'] = result2
```

```python
df_merged_final2.shape
```

```python
df_merged_final2['exclusion_flag'].value_counts(dropna = False)
```

```python
df_merged_final2['Exclusion'] = df_merged_final2['exclusion_flag'] == 'Low-activity customer'
```

```python
df_merged_final2_filtered = df_merged_final2[~df_merged_final2['Exclusion']]
```

```python
df_merged_final2_filtered = df_merged_final2_filtered.drop(columns=['Exclusion'])
```
